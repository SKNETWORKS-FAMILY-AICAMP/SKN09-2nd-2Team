### merge all of them
- msno

```python
merged_df = train_df

dfs_to_merge = [
    members_df,
    sample_sub_v2_df,
    sample_sub_zero_df,
    train_v2_df,
    transactions_v2_df,
    transactions_df
]

# 병합: msno, 중복 컬럼 구분을 위해 `suffixes` 추가
for i, df in enumerate(dfs_to_merge):
    merged_df = pd.merge(merged_df, df, on='msno', how='inner', suffixes=(f'_left{i}', f'_right{i}'))

# result
print(merged_df.head())
```
